**Problema 1: Hacer un programa que imprima los números del -100 al 200.**

Para que el programa muestre todos los números que van desde el -100 hasta el 200, utilicé un bucle **for** con la función **range(-100, 201)**, con la finalidad de que generara todos los números automáticamente. (Agregué el 201 ya que, de lo contrario, la serie solo mostraría hasta el 199).

```python
for i in range(-100, 201):
  print (i)
```

**Problema 2: Hacer un programa que imprima los números del -100 al 200 que sean divisibles entre 6 (con resto 0)**  

En este problema, también se muestran los números entre -100 y 200, pero con la condición de solo imprimir los que sean divisibles por 6. Para lograr esto, ocupé el operador **%**, de manera que únicamente genere los números cuyo resto sea 0, es decir, que el número sea divisible por 6.



```python
for i in range(-100, 201):
  if i % 6 == 0:
    print(i)
```

**Problema 3: Pseudocódigo, Sumador y comparador de dos números**

Acá utilicé las condiciones **if, elif y else** para aplicar más de una condición. En este caso, el programa suma dos números entregados por el usuario y comparara el resultado con los siguentes requisitos:


*   Si la suma es menor que 200, el programa indica que "Menor que 200".
*   Si está entre el rango de 200 a 250, se muestra el mensaje "Mayor que 200 y menor que 250".
*   Pero si el número sobrepasa los 250, el programa señala que la suma es "Mayor que 250".

```python
numero1 = 28
numero2 = 197
suma = numero1 + numero2
if suma < 200:
  print("Menor que 200")
elif suma > 200 and suma < 250:
  print("Mayor que 200 y menor que 250")
else:
  print("Mayor que 250")
```

**Problema 4: Pseudocódigo, Calculador de promedio**

Finalmente, trabajé con una función que revisa dos aspectos del usuario: la edad y si le gusta la música urbana. Si la persona tiene más de 18 años y además le gusta ese tipo de música, el programa entrega un mensaje específico afirmativo. En cualquier otro caso, se imprime un mensaje negativo. De esta manera, este problema maneja no solo condiciones, sino que también variables.

```python
edad = 21
gusta_musica_urbana = True
if edad > 18 and gusta_musica_urbana == True:
  print("Eres mayor de edad y te gusta la música urbana")
else:
  print("No eres mayor de edad o no te gusta la música urbana")
```
